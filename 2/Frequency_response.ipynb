```python
import numpy as np
import matplotlib.pyplot as plt
```

## Frequency response of PT1 term

The frequency response is

$$ G(j\omega) = \frac{A}{1+T_1 j\omega} $$

```python
# parameters
T1 = 10
A = 1
# data for Bode-plot
omega = np.logspace(-4,3,1000)
Gjom = A/(1+T1*1j*omega)
K = np.abs(Gjom)
phi = np.angle(Gjom)
```

```python
# plot
plt.figure(figsize=(8,6))
plt.plot(omega, K);
```

```python
# plot
plt.figure(figsize=(8,6))
plt.semilogx(omega, 20*np.log10(K));
```

```python
# Bode plot, magnitude
plt.figure(figsize=(8,6))
plt.semilogx(omega, np.log10(np.abs(Gjom)))
plt.xlabel(r'$\omega$'), plt.ylabel(r'$K(\omega)[dB]$')
plt.grid(True);
```

```python
# Bode plot, phase
plt.figure(figsize=(8,6))
plt.semilogx(omega, phi)
plt.xlabel(r'$\omega$'), plt.ylabel(r'$\varphi(\omega)[rad]$')
plt.grid(True);
```

```python
# Bode plot, amplidude and phase
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.semilogx(omega, np.log10(np.abs(Gjom)))
plt.ylabel(r'$K(\omega)[dB]$'), plt.grid(True)
plt.subplot(2,1,2)
plt.semilogx(omega, phi)
plt.xlabel(r'$\omega$'), plt.ylabel(r'$\varphi(\omega)[rad]$')
plt.grid(True);
```

```python
# data for Nyquist-plot
reG = np.real(Gjom)
imG = np.imag(Gjom)
```

```python
# Nyquist-plot
plt.figure(figsize=(8,6))
plt.plot(reG, imG)
plt.xlabel('Re'), plt.ylabel('Im')
plt.grid(True);
```

### FOPTD term (PT1 + Delay)

Frequency response:

$$ G(j\omega) = \frac{A}{1+T_1 j\omega} e^{-\tau j\omega} $$

```python
# parameters
tau = 1
T1 = 10
A = 1
# G(j omega)
omega = np.logspace(-4,3,1000)
Gjom = A/(1+T1*1j*omega)*np.exp(-tau*1j*omega)
K = np.abs(Gjom)
Kdb = 20*np.log10(K)
phi = np.angle(Gjom)
```

```python
# Bode plot, amplidude and phase
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.semilogx(omega, Kdb)
plt.ylabel(r'$K(\omega)[dB]$'), plt.grid(True)
plt.subplot(2,1,2)
plt.semilogx(omega, phi)
plt.xlabel(r'$\omega$'), plt.ylabel(r'$\varphi(\omega)[rad]$')
plt.grid(True);
```

Unwrapping phase

```python
# Bode plot, amplidude and phase
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.semilogx(omega, Kdb)
plt.ylabel(r'$K(\omega)[dB]$'), plt.grid(True)
plt.subplot(2,1,2)
plt.semilogx(omega, np.unwrap(phi))
plt.xlabel(r'$\omega$'), plt.ylabel(r'$\varphi(\omega)[rad]$')
plt.grid(True);
```

```python
# data for Nyquist-plot
reG = np.real(Gjom)
imG = np.imag(Gjom)
```

```python
# Nyquist-plot
plt.figure(figsize=(8,6))
plt.plot(reG, imG)
plt.xlabel('Re'), plt.ylabel('Im')
plt.grid(True);
```

### PT2 term

Transfer function:

$$ G(s) = \frac{A}{T^2s^2 + 2\xi Ts + 1} $$

```python
# Freq response, Bode data, Nyquist data
# parameters
A = 1
T = 10
xi = 0.4
# data
omega = np.logspace(-3,1,1000)
Gjom = A/(T**2*(1j*omega)**2 + 2*xi*T*1j*omega + 1)
K = np.abs(Gjom)
Kdb = 20*np.log10(K)
phi = np.angle(Gjom)
```

```python
# Bode plot, amplidude and phase
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.semilogx(omega, Kdb)
plt.ylabel(r'$K(\omega)[dB]$'), plt.grid(True)
plt.subplot(2,1,2)
plt.semilogx(omega, phi)
plt.xlabel(r'$\omega$'), plt.ylabel(r'$\varphi(\omega)[rad]$')
plt.grid(True);
```

```python
# Nyquist-plot
reG = np.real(Gjom)
imG = np.imag(Gjom)
# plot
plt.figure(figsize=(8,6))
plt.plot(reG, imG)
plt.xlabel('Re'), plt.ylabel('Im')
plt.grid(True);
```
