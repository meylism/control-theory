```python
import numpy as np
import matplotlib.pyplot as plt
```

## PT3 system with PI controller (FreqDomain)

The process is:

$$ P(s) = \frac{k}{(1 + T_1 s)(1 + T_2 s)(1 + T_3 s)}$$

the controller is:

$$ C(s) = K_p \left(1 + \frac{1}{T_i s} \right)$$

```python
# definition of the parameters and frequency responses
k = 2
T1 = 1
T2 = 1
T3 = 1

Kp = 2
Ti = 5

N = 1000
omega = np.logspace(-4,3,N)
Pjom = k/((1+T1*1j*omega)*(1+T2*1j*omega)*(1+T3*1j*omega))
Cjom = Kp*(1+1/(Ti*1j*omega))

# open loop
G0jom = Pjom*Cjom
K0 = np.abs(G0jom)
K0_dB = 20*np.log10(K0)
phi0 = np.unwrap(np.angle(G0jom))
reG0 = np.real(G0jom)
imG0 = np.imag(G0jom)
```

```python
# Bode plots
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.semilogx(omega,K0_dB), plt.grid(True)
plt.xlabel('$\omega$'), plt.ylabel('$K_0(\omega)$')
plt.semilogx(omega,0*K0_dB,'r:')
plt.subplot(2,1,2)
plt.semilogx(omega,phi0), plt.grid(True)
plt.semilogx(omega,0*phi0 - np.pi,'r:')
plt.xlabel('$\omega$'), plt.ylabel('$\phi_0(\omega)$');
```

```python
# Nyquist plot
plt.figure(figsize=(8,6))
plt.plot(reG0, imG0), plt.grid(True)
plt.ylim([-10,1]);
```
